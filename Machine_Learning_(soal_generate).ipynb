```python
import pandas as pd
from datasets import Dataset
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Path file CSV di Google Drive
file_path = '/content/drive/My Drive/Dataset Soal/soal_bahasa_indonesia_pilihan_ganda.csv'

# Memuat dataset menggunakan pandas
data = pd.read_csv(file_path)

# Convert to Huggingface Dataset
dataset = Dataset.from_pandas(data)

# Inisialisasi tokenizer dan model
tokenizer = T5Tokenizer.from_pretrained('t5-small')
model = T5ForConditionalGeneration.from_pretrained('t5-small')

def preprocess_function(examples):
    inputs = [f"generate question: {materi}" for materi in examples['materi']]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding='max_length')

    labels = tokenizer(examples['soal'], max_length=512, truncation=True, padding='max_length')['input_ids']
    model_inputs['labels'] = labels
    return model_inputs

# Preprocess the dataset
tokenized_dataset = dataset.map(preprocess_function, batched=True)
```

```python
from transformers import Trainer, TrainingArguments

# Definisikan argumen pelatihan
training_args = TrainingArguments(
    output_dir='./results',
    eval_strategy="epoch",  # Menggunakan eval_strategy yang benar
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

# Inisialisasi Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset,
)

# Melatih model
trainer.train()
```

```python
# Evaluasi model
eval_results = trainer.evaluate()

print(f"Evaluation results: {eval_results}")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load tokenizer and model T5
tokenizer = T5Tokenizer.from_pretrained('t5-small')
model = T5ForConditionalGeneration.from_pretrained('t5-small')

def generate_question_from_material(material_text):
    # Format input for model T5
    inputs = f"generate question: {material_text}"
    input_ids = tokenizer.encode(inputs, return_tensors='pt', max_length=512, truncation=True)

    # Generate output from model T5
    outputs = model.generate(input_ids, max_length=150, num_beams=4, early_stopping=True)

    # Decode output
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract question from generated text
    question = process_generated_text(generated_text)

    return question

def process_generated_text(generated_text):
    # Split generated text into question and options
    parts = generated_text.split('?')
    question = parts[0] + '?'

    return question.strip()

def generate_options(material_text):
    # Generate options from material text
    words = material_text.split()
    options = words[:4]  # Take first four words as options

    # Format options as A, B, C, D
    formatted_options = [f"A. {options[0]}", f"B. {options[1]}", f"C. {options[2]}", f"D. {options[3]}"]

    return formatted_options

def generate_correct_answer(material_text):
    # Generate correct answer from material text
    words = material_text.split()
    correct_answer = words[0]

    return correct_answer

def generate_complete_question(material_text):
    # Generate complete question
    question = generate_question_from_material(material_text)
    options = generate_options(material_text)
    correct_answer = generate_correct_answer(material_text)

    # Construct complete question format
    complete_question = {
        "question": question,
        "options": options,
        "answer": correct_answer
    }

    return complete_question

def main():
    # Prompt user for material text
    material_text = input("Masukkan teks materi: ")

    # Generate complete question
    generated_question = generate_complete_question(material_text)

    # Print generated question, options, and correct answer
    print("\nGenerated Question:")
    print("Question:", generated_question['question'])
    print("Options:")
    for option in generated_question['options']:
        print(option)
    print("Answer:", generated_question['answer'])

if __name__ == "__main__":
    main()
```
